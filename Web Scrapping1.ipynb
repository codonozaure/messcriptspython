```python
#structure url2
import requests
page= requests.get("http://www.pythonscraping.com/pages/warandpeace.html")
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all('span',{"class": "green"}, limit=5)
```

```python
soup.find_all('span',{"class":"red"}, limit=5)
```

```python
soup.find_all('span',{"class": "green"}, limit=1)
```

```python
soup.find_all('span',{"class":"red"}, limit=1)
```

```python
#soit en vert soit en rouge
soup.find_all('span',{'class':['green','red']},limit=1)
```

```python
#the prince en miniscule coloré en vert 
soup.body.findAll(text='the prince',attrs={'class':'green'})
```

```python
#identifier la structure de la page web URL3!
import requests
page= requests.get("http://www.pythonscraping.com/pages/page3.html")
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
#gestion des erreurs 
import requests
from urllib.error import HTTPError, URLError
from bs4 import BeautifulSoup
try:
    page= requests.get("https://twitter.com/realDonalTrump")
    soup = BeautifulSoup(page.content, 'html.parser')

except HTTPError:
   print(HTTPError)
except URLError:
   print('ERROR404')
else:
   print('ok')
```
